```python
import re
import fandom
from collections import defaultdict
fandom.set_wiki("onepiece")
```

```python
char_appearances = defaultdict(set)
```

```python
def process_episodes(start, end):
  for i in range(start, end):
    page = fandom.page(title=f"Episode {i}")
    chars_str = page.content["sections"][2]["content"]
    chars = chars_str.split("\n")

    for char in chars:
      match = re.match(r"([^()]+)", char)
      if match:
        char_name = match.group(1).strip()
        char_appearances[char_name].add(i+1)
```

```python
def shared_episodes(chars):
  if len(chars) < 2:
    raise ValueError("Must have at least 2 characters.")

  res = char_appearances[chars[0]]

  for i in range(1, len(chars)):
    res = res.intersection(char_appearances[chars[i]])

  return res
```

```python
shared_episodes(["Monkey D. Luffy", "Donquixote Doflamingo"])
```

```python
keys_to_remove = ['↑ One Piece Film: Strong World Episode 0, MyAnimeList', '↑ One Piece Episode 0 Manga to be Animated, AnimeNewsNetwork', '↑ ONE PIECE Official YouTube Channel', '↑ Eiichiro_Staff on Twitter, the English subtitles are announced.']

for key in keys_to_remove:
  del char_appearances[key]

char_appearances.keys()
```

```python
nodes = []
for char in char_appearances.keys():
  nodes.append({"id": char})

display(nodes)
```

```python
edges = []
visited = set()

for char in char_appearances.keys():
  for related_char in char_appearances.keys():
    if char != related_char and char not in visited:
      num_eps = len(shared_episodes([char, related_char]))
      if num_eps > 0:
        edges.append({"source": char, "target": related_char, "value": num_eps})

  visited.add(char)

display(edges)
```
